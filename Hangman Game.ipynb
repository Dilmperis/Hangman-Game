## Hangman game 

```python
import random as rd

# Choosing a random word from the list:
word_list = ["apple", "banana", "orange", "grape", "pineapple", "kiwi", "watermelon", "strawberry"]
word = rd.choice(word_list)

# Count the letters and set an equivalent "empty list":
len_word = len(word)
letters = list(word)
empty_list = ["_" for _ in range(len_word)]
print(f'Guess a letter, for the word: {empty_list} ')

wrong_letter = 0  # Count how many wrong letters are chosen
extra_lives = 10

while wrong_letter < extra_lives:
    print('Give a letter: \n')
    letter = input().lower()  # Convert input to lowercase
    if len(letter) != 1 or not letter.isalpha():
        print("Please enter a single letter.")
        continue

    if letter in letters:
        indices = [index for index, char in enumerate(letters) if char == letter]
        for ind in indices:
            empty_list[ind] = letter
        print(empty_list)
        if "_" not in empty_list:
            print(f"You WIN! The word was {word}")
            break
    else:
        wrong_letter += 1
        print(f'Wrong letter! You have {extra_lives - wrong_letter} lives left.')

if wrong_letter >= extra_lives:
    print(f'\nYou Lost, the word was {word}')
```
